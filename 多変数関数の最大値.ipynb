現在，高校3年生の数学の演習の授業を担当しています。その中で扱った問題です。

## 問題

$a$ と $b$ は実数とし，関数 $f(x)=x^2+ax+b$ の $0\leqq x\leqq 1$ における最小値を $m$ とする。

1. $m$ を$a$ と $b$ で表せ。
2. $a+2b\leqq 2$ を満たす $a$ と $b$ で $m$ を最大にするものを求めよ。また，このときの $m$ の値を求めよ。

-----

## 解答（前半）

$f'(x)=2x+a$ なので，

- $-\frac{a}2\leqq 0$，すなわち，$a\geqq 0$のとき，$$m=f(0)=b$$
- $0\leqq -\frac{a}2\leqq 1$，すなわち，$-2\leqq a\leqq 0$のとき，$$m=f\left(-\frac{a}2\right)=-\frac{a^2}4+b$$
- $-\frac{a}2\geqq 1$，すなわち，$a\leqq -2$のとき，$$m=f(1)=1+a+b$$

これを $abm$ 空間で考えて図示してみよう！

```julia
using PlotlyJS

function f(a,b)
    if a <= -2
        a+b+1
    elseif a <= 0
        b-a*a/4
    else
        b
    end
end

x = -5:.1:5
y = -5:.1:5
z1 = [f(i,j) for i in x, j in y]'

# layout = Layout(
#     title="多変数関数のグラフ",
#     autosize=false,
#     width=500,
#     height=500,
#     margin=attr(l=65, r=50, b=65, t=90)
# )

plot(surface(z=z1))
```


## 解答（後半）

 $a+2b\leqq 2$ なので，まず$a$を固定して考えてみます。
すると，$b$の範囲は

$$b\leqq \frac{2-a}2\cdots(1)$$

となります。

 - $a\geqq 0$ で　$a$　が固定されているとき，$m=b$ です。

$b$　のとりうる範囲は(1)なので，　
$$m=b\leqq \frac{2-a}2$$

となります。次に，$a$ を変化させます。$a$を変化させる範囲は $a\geqq 0$ ，$a=0$で $\frac{2-a}2$ は最大となるので，

$$m=b\leqq \frac{2-a}2\leqq \frac{2-0}2=1$$

$$\therefore m\leqq 1$$

となります。

 - $-2\leqq a\geqq 0$ で　$a$　が固定されているとき，$m=-\frac{a^2}4+b$ です。

$b$　のとりうる範囲は(1)なので，　

$$m=-\frac{a^2}4+b\leqq-\frac{a^2}4+\frac{2-a}2=-\frac14\left(a+1\right)^2+\frac54$$

となります。次に，$a$ を変化させます。$a$を変化させる範囲は $-2\leqq a\geqq 0$ ，$a=-1$で $-\frac14\left(a+1\right)^2+\frac54$ は最大となるので，

$$m=-\frac{a^2}4+b\leqq-\frac{a^2}4+\frac{2-a}2=-\frac14\left(a+1\right)^2+\frac54\leqq\frac54$$

$$\therefore m\leqq \frac54$$

となります。



 - $a\leqq -2$ で　$a$　が固定されているとき，$m=1+a+b$ です。

$b$　のとりうる範囲は(1)なので，　

$$m=1+a+b\leqq1+a+\frac{2-a}2=\frac12a+2$$

となります。次に，$a$ を変化させます。$a$を変化させる範囲は $a\geqq -2$ ，$a=-2$で $\frac12a+2$ は最大となるので，

$$m=1+a+b\leqq1+a+\frac{2-a}2=\frac12a+2\leqq \frac12\cdot(-2)+2=1$$

$$\therefore m\leqq 1$$

以上より，$\bm{\max m=\frac54 (a=-1,\,b=\frac32)}$となります。

```julia
function g(a,b)
if a+2b <= 2
    f(a,b)
else
    0
end
end


x = -5:0.1:5
y = -5:0.1:5
z2 = [g(i,j) for i in x, j in y]' 

# layout = Layout(
#     title="多変数関数の最大値その2",
#     autosize=false,
#     width=500,
#     height=500,
#     margin=attr(l=65, r=50, b=65, t=90)
# )

plot(surface(z=z2))
```
